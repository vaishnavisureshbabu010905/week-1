```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your CSV data
df = pd.read_csv('solar.csv')  # Ensure the correct file path

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)  # Note 'sns.histplot' is used
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.title('Distribution of Generated Power')
plt.show()

```

```python
df[df.columns[18:20]].hist(bins=30,figsize=(4,6))
plt.show()
```

```python
#Bivariate analysis
#Scatter plot with target feature
plt.figure(figsize=(15,30))
for i,column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column],df['generated_power_kw'])
    plt.title(f'{column} vs Generated Power(kW)')
    plt.xlabel(column)
    plt.ylabel('Generated Power (kW)')
plt.tight_layout()
plt.show()    
```
